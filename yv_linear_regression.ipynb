# Linear Regression Model
## For Predicting Money Gived For n Of Views


problem statement : given number of views n try to predict money that you will get for those views 
sample training example : youtube gives 20$ dollars for each 1000 views, money per each 1k views Training Example : nk views as input feature for n\$ dollars as target variable

# Training set
| Views  | Money |
|-------|-----|
| 1.0 | 25.0  | 
| 3.0    | 75.0  | 
| 6.5  | 162.5  | 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define our training set

# input features
x_train = np.array([1.0,3.0,6.5])

# target variables
y_train = np.array([25.0,75.0,162.5]) 

# number of training examples
m = x_train.shape[0]

print(f"number of training examples is: {m}")
# training example x^i y^i
i = 0
print(f"y^(i) , x^(i)")
print(f"{[x_train[i],y_train[i]]}")
```

# Ploting The Data



```python
plt.scatter(x_train,y_train,marker="x",c="r")
plt.title("Youtube Baying System")
plt.xlabel("Views in k's")
plt.ylabel("Money in dollars")
```

## Model Construction


```python
# our weights , coeffictions, model parameters
w = 200
b = 100
```

```python
# model function

def model_output(x,w,b) :
    m = x.shape[0]

    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

## Testing The Model Output : staright line that fits the data 

```python
f_wb = model_output(x_train,w,b)
```

```python
plt.plot(x_train,f_wb,c="b",label="Our Prediction")

plt.scatter(x_train,y_train,marker="x",c="r",label="Actual Values")
plt.xlabel("Views in k's")
plt.ylabel("Money in dollars")


plt.legend()
plt.show()
```

### Cost Function

```python
def cost_func(x,y,w,b):

    m = x.shape[0]

    cost_sum = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) **2
        cost_sum = cost_sum + cost 
        total_cost = (1 /( 2 * m)) * cost_sum 
    return total_cost
    
```
